```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score 
from sklearn.metrics import confusion_matrix
```

```python
df = pd.read_csv("train.csv")
df.head()
```

```python
dfknn = df.drop(['Name','PassengerId','Ticket','Cabin'],axis=1)
dfknn.head()
```

```python
le = LabelEncoder()
dfknn['Sex'] = le.fit_transform(dfknn['Sex'])
dfknn['Embarked'] = le.fit_transform(dfknn['Embarked'])
```

```python
def knn_data(kd):
    x = dfknn.drop(['Pclass'],axis=1)
    y = dfknn['Pclass']
    a = []
    b = []
    
    for i in range(1,kd):
        knn = KNeighborsClassifier(n_neighbors = i)
        a.append(i)

        print(f" \nHere the K value is: {i} ")
              
        x_train,x_test,y_train,y_test = train_test_split(x,y, test_size = 0.2, random_state = 10)
        model = knn.fit(x_train,y_train)
        pred = model.predict(x_test)
              
        score = model.score(x_test,y_test)
        b.append(score)
              
        print(f"\nThe Accuracy score when K is {i} ", score)
        print(f"Confusion Matrix when K is {i} \n", confusion_matrix(y_test, pred))
        
              
    plt.plot(a,b)
    plt.show()
```

```python
knn_data(268)
```
