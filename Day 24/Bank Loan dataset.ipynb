```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.tree import DecisionTreeClassifier,export_graphviz
from sklearn.ensemble import RandomForestClassifier
```

```python
df = pd.read_excel("Bank_Personal_Loan_Modelling.xlsx",sheet_name=1)
df.head()
```

```python
df.isnull().sum()
```

```python
df.describe().T
```

```python
df = df.drop(['ID','ZIP Code'], axis=1)
```

```python
x = df.drop(['Personal Loan'], axis=1)
y = df['Personal Loan']
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state = 10)
```

```python
RandomForrest = RandomForestClassifier(n_estimators=1000, max_features=2, oob_score=True)
```

```python
RandomForrest.fit(x_train, y_train)
```

```python
for feature, imp in zip(x_train.columns, RandomForrest.feature_importances_):
    print(f"{feature} {imp}")
```

```python
features = []
for feature, imp in zip(x_train.columns, RandomForrest.feature_importances_):
    if imp >= 0.05:
        features.append(feature)
```

```python
features
```

```python
tree_model = DecisionTreeClassifier(random_state=101)
tree_model.fit(x_train[features], y_train)
```

```python
tree_model.score(x_train[features], y_train)
```

```python
print("The accuracy on test data is: ", tree_model.score(x_test[features], y_test))
```

```python
with open("Loan_tree.dot",'w') as f:
    f = export_graphviz(tree_model, feature_names=features, out_file=f)
```
